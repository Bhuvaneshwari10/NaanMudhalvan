```python
#Outlier median
q1 = df.lot_area_renov.quantile(0.25)   #(Q1)
q3 = df.lot_area_renov.quantile(0.75)   #(Q3)
```

```python
print(q1)
print(q3)
```

```python
IQR=q3-q1
```

```python
Upper_limit = q3+1.5 * IQR
lower_limit = q1-1.5 * IQR
```
